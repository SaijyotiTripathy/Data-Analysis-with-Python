## **Indexing, Selection and Slicing**

Individual columns in the DataFrame can be selected with regular indexing. Each column is represented as a Series

```python
df
```

```python
# loc selects rows matching the given index
df.loc[['France','Japan']]
```

```python
# iloc works with the (numeric) "position" of the index
df.iloc[-1]
```

```python
# selects columns
df['Population']
```

```python
# converts to DataFrame
df['Population'].to_frame()
```

```python
df[['Population','GDP']]
```

```python
# direct slicing
df[2:4]
```

```python
df.loc['France':'Italy']
```

As a second "argument", you can pass the column(s) you'd like to select

```python
df.loc['France':'Italy','Population'].to_frame()
```

```python
df.loc['France':'Italy',['Population','GDP']]
```

```python
df
```

```python
df.iloc[3]
```

```python
df.iloc[[0,1,-1]]
```

```python
df.iloc[1:4]
```

```python
df.iloc[1:4,3]
```

```python
df.iloc[1:4, [0,1]]
```

```python
# rows 1,2,3 and columns starting from 1 excluding last 
df.iloc[1:4, 1:-1]
```

### **Always use loc and iloc to reduce ambiguity, specially with DataFrames with numeric indexes.**

## **Dropping stuff**

Instead of pointing out which values you'd like to **select** you point which ones you'd like to **drop**

All these drop methods return a new DataFrame. Doesn't change the original DataFrame (immutable)

```python
# drop by index
df.drop('Canada')
```

```python
# drop by indices
df.drop(['Canada','Italy'])
```

```python
# drop by columns
df.drop(columns= ['Population','GDP'])
```

**axis=0**: horizontal axis (rows) <br>
**axis=1** : vertical axis (columns)

```python
df.drop(['Italy','Canada'], axis=0)
```

```python
df.drop(['Italy','Canada'], axis='rows')
```

```python
df.drop(['Population','GDP'], axis=1)
```

```python
df.drop(['Population','GDP'], axis='columns')
```

## **Modifying DataFrames**

### Adding new column

```python
langs= pd.Series(['French','German','Italian'], index=['France','Germany','Italy'], name='Language')
```

```python
langs
```

```python
df['Language']= langs
```

```python
df
# NaN : blank 
```

### Replacing values per column

```python
df['Language']= 'English'
```

```python
df
```

### Renaming columns

```python
# immutable operation
df.rename(columns= {'HDI': 'Human Development Index','APC': 'Annual Popcorn Consumption'}, index={'United States': 'USA', 'United Kingdom':'UK'})
```

```python
df.rename(index=str.upper)
```

```python
df.rename(index=lambda x: x.lower())
```

```python
df
```

### Dropping columns

`inplace` attribute is used to modify the original DataFrame

```python
df
```

```python
df.drop(columns='Language', inplace=True)
```

```python
df
```

### Adding values

Append returns a new DataFrame

```python
df.append(pd.Series({'Population':3, 'GDP':5}, name='China'))
```

```python
df
```

Directly sets the new index and values to the DataFrame

```python
df.loc['China']= pd.Series({'Population': 1_400, 'Continent':'Asia'})
```

```python
df
```

Drops the row by index 

```python
df.drop('China', inplace=True)
```

```python
df
```
